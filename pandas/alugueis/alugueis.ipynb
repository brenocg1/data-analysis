```python
list(dados["Tipo"].drop_duplicates())
```

```python
# Selecionando apenas os residencias
residencial  = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa de Vila',]
residencial
```

```python
# Vou selecionar dentro do dataframe apenas os imóveis residencias agora
dados.head(10)
```

```python
dados = pd.read_csv("../datasets/aluguel.csv", sep = ';')
selecao = dados["Tipo"].isin(residencial)
selecao.head(10)
```

```python
# Agora tenho um DataFrame apenas com imóveis residencias, que é o que me interessa
dados_residencial = dados[selecao]
dados_residencial.head(10)
```

```python
print("Quantidade de imóveis do DataFrame original %.2f" % dados.shape[0])
print("Quantidade de imóveis do DataFrame com apenas residencias %.2f" % dados_residencial.shape[0])
```

```python
# Recriando os indices do novo dataframe
dados_residencial.index = range(dados_residencial.shape[0])
dados_residencial.head(10)
```

## Exportando base de dados

```python
# dessa maneira ele tbm vai importar os indices
# como uma nova coluna no .csv 
# para isso passei index = false
dados_residencial.to_csv("../datasets/dados_residencial.csv", sep  = ';', index = False)
```

## Seleções Residênciais

### Extraindo algumas info desse DataFrame

```python
# Selecionar imóveis classificados como Apartamento
selecao = dados_residencial['Tipo'] == "Apartamento"
dados_aps = dados_residencial[selecao].shape[0]
print("%d Apartamentos" % dados_aps)
```

```python
# Selecionar imóveis como tipo Casa, Casa de Condomínio e Casa de Vila
selecao = (dados_residencial["Tipo"] == "Casa") | (dados_residencial["Tipo"] == "Casa de Condomínio") | (dados_residencial["Tipo"] == "Casa de Vila")
print("Quantidade de imóveis desse tipo Casa, Casa de Condomínio ou  Casa de Vila %d" % dados_residencial[selecao].shape[0]) 
```

```python
# Selecionar imóveis entre 60 e 100 metros quadrados
selecao = (dados_residencial['Area'] >= 60) & (dados_residencial['Area'] <= 100)
print("Quantidade de imóveis na faixa entre 60 e 100 m2 %d" % dados_residencial[selecao].shape[0])
```

```python
# Selecionar imóveis com pelo menos 4 quartos e aluguel menor que R$2000,00
selecao = (dados_residencial["Quartos"] >= 4) & (dados_residencial["Valor"] < 2000)
print("Quantidade de imóveis com menos de 4 quartos e aluguel menor que 2000: %d" % dados_residencial[selecao].shape[0])
```
