## Identificando e Removendo Outliers

```python
%%html
<img src = "../imgs/Boxplot-04.png", width = 2000, height = 1000 >
```

```python
# A imagem acima serve para ajudar a identificar os outliers pelo Boxplot
# Calculando Quantis 
valor = dados["Valor"]
Q1 = valor.quantile(.25)
Q3 = valor.quantile(.75)
IQR = Q3 - Q1
limite_inf = Q1 - 1.5 * IQR
limite_sup = Q3 + 1.5 * IQR

# Selecionando dados apenas dentro desses limites, quem ta fora é outliers...
selecao = (valor >= limite_inf) & (valor <= limite_sup)
dados_new = dados[selecao]

dados_new.boxplot(["Valor"])
```

```python
# diferença na hora de fazer o histograma com os outliers e sem eles
dados.hist("Valor")
```

```python
# Sem outliers
dados_new.hist("Valor")
```

```python
# Removendo outliers baseado nos grupos
dados.boxplot(["Valor"], by = ["Tipo"])
# Pois misturando esses tipos, podemos facilmente ter valores de aluguel alto para uma Casa de Condomínio,
# diferente de um Quitinete
```

```python
grupo_tipo = dados.groupby('Tipo')['Valor']
# fazendo os calculos anteriores só que para esse tipo somente, e com a categoria Valor, obviamente
Q1 = grupo_tipo.quantile(.25)
Q3 = grupo_tipo.quantile(.75)
IQR = Q3 - Q1
limite_inf = Q1 - 1.5 * IQR
limite_sup = Q3 + 1.5 * IQR
```

```python
dados_empty = pd.DataFrame()

for tipo in grupo_tipo.groups.keys():
    is_tipo = dados['Tipo'] == tipo
    is_dentro = (dados["Valor"] >= limite_inf[tipo]) & (dados["Valor"] <= limite_sup[tipo])
    selecao = is_dentro & is_tipo
    dados_selecao = dados[selecao]
    dados_empty = pd.concat([dados_empty, dados_selecao])
```

```python
dados_por_tipo = dados_empty
dados_por_tipo.boxplot(["Valor"], by = ["Tipo"], widths = (.5, .5, .5, .5, .5), patch_artist = True)
```
